3:Check the Integar Value is Odd or Even?

```python
Number=int(input("Enter the any Integar Number:"))
if(Number%2!=0):
  print(f"Your Number {Number} is Odd")
else:
  print(f"Your Number {Number} is Even")
```

4:Check the Leap Year?

```python
Year=int(input("Enter the Year:"))
if(Year%4==0):
  print(f"{Year} is Leap Year")
else:
  print(f"{Year} is Not Leap Year")
```

5:Check the Word is Vowels Or Not?

```python
word=input("Enter the Any Word:")
if(word=="a","e","i","o","u"):
  print(f"{word} is Vowels")
elif(word=="A","E","I","O","U"):
  print(f"{word} is Vowels")
else:
  print(f"{word} is Not Vowels")
```

#*For Loop*

1:Print The value of 2 1 4 3 6 5 8 7

```python
for a in range(1,9,2):
  print(a+1,end=" ")
  print(a,end=" ")
```

2:Calculate the Sum of n number

```python
n=int(input("Enter the Any Number:"))
sum=0
for a in range(1,n+1):
   sum+=a
   print(a,end="+"if a<n else "" )
print(f"={sum}")
```
